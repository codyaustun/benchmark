```python
import os
import json
from glob import glob

import pandas as pd

from jinja2 import Template
```

```python
source = '../results/CIFAR10/train/'

paths = glob(os.path.join(source, "*.json"))
submissions = []
for path in paths:
    with open(path) as file:
        record = json.load(file)
        submissions.append(record)
        
submissions = sorted(submissions, key=lambda x: x['time'])
for index, submission in enumerate(submissions):
    submission['rank'] = index + 1
submissions
```

```python
html = """
<!--lorem-->
<p>Duis ac erat. Etiam consequat. Integer sed est eu elit pellentesque dapibus.
Duis venenatis magna feugiat nisi. Vestibulum et turpis. Maecenas a enim.
Suspendisse ultricies ornare justo. Fusce sit amet nisi sed arcu condimentum
venenatis. Vivamus dui. Nunc accumsan, quam a fermentum mattis, magna sapien
iaculis pede, at porttitor quam odio.</p>
<!--/lorem-->

    
<div class="table-responsive">
    <table class="table">
        <thead>
            <tr>
                <th>Rank</th>
                <th>Model</th>
                <th>Time (in hours)</th>
                <th>Accuracy</th>
            </tr>
        </thead>
        <tbody>
        {% for submission in submissions %}
            <tr>
                <td>
                    {{submission.rank}}
                </td>
                <td>
                    {{submission.model}}
                </td>
                <td>{{submission.time}}</td>
                <td>{{submission.accuracy}}%</td>
            </tr>
        {% endfor %}
        </tbody>
    </table>
</div>

"""
```

```python
with open('../templates/CIFAR10/train.html') as file:
    print(file.read())
```

```python
print(template.render(submissions=submissions))
```
